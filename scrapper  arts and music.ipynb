```python
import time
from bs4 import BeautifulSoup, SoupStrainer
import requests
import csv
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

url ="https://www.youtube.com/results?search_query=art+and+music+of+america"
browser = webdriver.Chrome(executable_path=r'C:/Users/ashish saha/.wdm/chromedriver/2.46/win32/chromedriver.exe')

browser.get(url)
time.sleep(1)
csv_file = open('YouTube arts1.csv','w',encoding='utf-8')
elem = browser.find_element_by_tag_name("body")
n=1

total_videos = 100
while total_videos:
    no_of_pagedowns = 20
    while no_of_pagedowns:
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.2)
        no_of_pagedowns-=1
        
    source = requests.get(url).text
    container = SoupStrainer(class_="yt-lockup-content")
    soup = BeautifulSoup(source, 'lxml', parse_only=container)

    csv_writer = csv.writer(csv_file)
    
    for content in reversed(soup.select('.yt-lockup-content')):
        try:
            title = content.h3.a.text
            print(title)

            id = content.h3.a.get('href')
            id = "https://www.youtube.com" + id
            print(id)

            description = content.find('div', class_="yt-lockup-description yt-ui-ellipsis yt-ui-ellipsis-2").text
            print(description)
        
        except Exception as e:
            description = None

        csv_writer.writerow([n,id,title,description])
        
        n=n+1
        time.sleep(2)
        total_video = total_videos-20

csv_file.close()
```
